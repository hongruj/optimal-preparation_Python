```python
import numpy as np
from lib.simple import Sim
from lib.plotting import set_figure, plot_fr_per_cond
from lib.defaults import Main
from lib.plotting import plot_out
```

```python
Simple = Sim()
spontaneous, xstars, c = Simple.read_setup()
spontaneous = spontaneous[:,None]
n_mov = xstars.shape[1]
q = Simple.q_mov(c) 
r, t_move = Simple.vanilla(q, xstars,spontaneous)
```

```python
r = np.array(r)
axes = set_figure(n_mov)  
for j in range(n_mov):
    plot_fr_per_cond(axes, j, t_move, 20*r[:,:,j], xstars)
```

```python
gamma = Main().gamma

out = [c @ gamma @ r[:,:,i].T for i in range(n_mov)]
out = np.array(out)
plot_out(out)
```
