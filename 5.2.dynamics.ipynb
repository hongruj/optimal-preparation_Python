```python
import numpy as np
from lib.simple import Sim
from lib.plotting import set_figure, plot_fr_per_cond, plot_out
from lib.defaults import Main
```

```python
Simple = Sim()
spontaneous, xstars, c = Simple.read_setup(inst=False)
spontaneous = spontaneous[:,None]
n_mov = xstars.shape[1]
r, _,_, t_move = Simple.dynamic(xstars,spontaneous)
```

```python
r = np.array(r)
axes = set_figure(n_mov)  
for j in range(n_mov):
    plot_fr_per_cond(axes, j, t_move, 20*r[:,:,j], xstars)
```

```python
gamma = Main().gamma

out = [c @ gamma @ r[-600:,:,i].T for i in range(n_mov)]
out = np.array(out)
plot_out(out)
```
